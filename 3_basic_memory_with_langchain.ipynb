```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", api_key=my_variable)
```

```python
prompts=[
    {"role":"user","content":"Hi my name is rahul chavan"},
    {"role":"ai", "content":"hello rahul chavan how can i asssit you today?"},
    {"role":"user","content":"what is name?"}
]
```

```python
res = llm.invoke(prompts)
```

```python
print(res.content)
```

```python
history = [] ## [{"role":"user", "content":"Who is pm of india ?"}]
while True:
    query = input("User: ")
    if query.lower() in ["exit", "quit", "bye"]:
        print("Good Bye 👋")
        break

    history.append({"role":"user", "content":query})
    print("User: ", query)


    res = llm.invoke(history)
    history.append({"role":"ai", "content":res.content})
    print("AI: ", res.content, "\n")
```
